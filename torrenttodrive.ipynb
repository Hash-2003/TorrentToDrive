```python
from pathlib import Path

Path(ARIA2_SESSION).touch(exist_ok=True)

session_text = Path(ARIA2_SESSION).read_text(errors="ignore").strip()

if session_text:
    print("Resuming from session file...")
    cmd = rf"""
    aria2c \
      --continue=true \
      --seed-time=0 \
      --enable-dht=true \
      --bt-enable-lpd=true \
      --bt-save-metadata=true \
      --max-connection-per-server=16 \
      --bt-max-peers=55 \
      --split=16 \
      --min-split-size=5M \
      --piece-length=1M \
      --file-allocation=none \
      --dir="{DATA_DIR}" \
      --input-file="{ARIA2_SESSION}" \
      --save-session="{ARIA2_SESSION}" \
      --save-session-interval=60 \
      --log="{ARIA2_LOG}" \
      --log-level=notice
    """
else:
    print("Starting fresh from magnet (seeding session)...")
    cmd = rf"""
    aria2c \
      --continue=true \
      --seed-time=0 \
      --enable-dht=true \
      --bt-enable-lpd=true \
      --bt-save-metadata=true \
      --max-connection-per-server=16 \
      --bt-max-peers=55 \
      --split=16 \
      --min-split-size=5M \
      --piece-length=1M \
      --file-allocation=none \
      --dir="{DATA_DIR}" \
      --save-session="{ARIA2_SESSION}" \
      --save-session-interval=60 \
      --log="{ARIA2_LOG}" \
      --log-level=notice \
      "{MAGNET_LINK}"
    """

!bash -lc "$cmd"
```
